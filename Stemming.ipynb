## Stemming

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. stemming is important in natural language understanding(NLU) and natual language processing(NLP).

## PorterStemmer

```python
words=["eating","eats","eaten","writing","writes","programming","programs","finally","finalize","history"]
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem('congratulation')
```

```python
stemming.stem('sitting')
```

## RegexStemmer class


```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemer.stem('eating')
```

```python
reg_stemer.stem('ingeating')
```

## Snowball Stemmer

```python
from nltk.stem import SnowballStemmer 
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

## Wordnet Lemmatization

Lemmatization technique is like stemming. the output we will get after lemmatization is called 'lemma'. which is a root word stem, the output of stemming, after lemmatization, we will be getting a valid word that means the same things.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. this class uses morphy() function to the wordNet CorpusReader class to find a lemma. Let us understand it with an example -

```python
## Q&A, chatbots, text summarization
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
''''
POS- Noun-n
verb-v
adjactive-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='v')
```

```python
''''
POS- Noun-n
verb-v
adjactive-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='n')
```

```python
''''
POS- Noun-n
verb-v
adjactive-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='a')
```

```python
''''
POS- Noun-n
verb-v
adjactive-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='r')
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","finally","finalize","history"]
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='n'))
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='a'))
```

```python
for word in words:
    print(word+"---->"+lemmatizer.lemmatize(word,pos='v'))
```

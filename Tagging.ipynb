## Parts of Speech Tag

### POS Tag Reference
CC → coordinating conjunction
CD → cardinal digit
DT → determiner
EX → existential there
FW → foreign word
IN → preposition/subordinating conjunction
JJ → adjective (beautiful)
NN → noun, singular
NNS → noun, plural
NNP → proper noun, singular (Taj, Mahal, Monument)
VBZ → verb, 3rd person singular present (is)
PRP → personal pronoun (I, he, she)

```python
#M.Sc. Chemistry Student Scholarship Scheme in India
paragraph = """"The M.Sc. Chemistry student scholarship schemes in India aim to support postgraduate learners in pursuing advanced studies in chemical sciences. These scholarships are offered by government bodies, universities, and private organizations to encourage research and academic excellence. The CSIR NET Fellowship is one of the most prominent schemes, providing financial aid to students qualifying for research. Similarly, the INSPIRE Scholarship by the Department of Science and Technology supports meritorious students in science streams. Many universities also provide merit-based scholarships to top-performing chemistry students. State governments run special schemes for economically weaker sections to ensure equal access to higher education. Private foundations and industries linked to chemical research also sponsor research grants. These scholarships not only reduce financial burden but also motivate students to focus on innovation and discovery. By supporting M.Sc. chemistry students, India strengthens its scientific workforce. Ultimately, such schemes contribute to national progress in pharmaceuticals, materials science, and sustainable chemistry."""
```

```python
import nltk

from nltk.corpus import stopwords

sentences=nltk.sent_tokenize(paragraph)
```

```python
## we will find the pos tag
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words('english'))]
    # sentences[i]=' '.join(words) # converting all the list of into sentences
    pos_tag=nltk.pos_tag(words)
    print(pos_tag)
```

## Stopwords with NLTK

```python
## Key Treaties and MoUs with Neighbours
paragraph = """"India, located in South Asia, is bounded by the Himalayas in the north and the Indian Ocean in the south, sharing borders with Pakistan, China, Nepal, Bhutan, Bangladesh, Myanmar, and Afghanistan. Its geography has made relations with neighbours both strategic and sensitive. With Pakistan, India signed the Indus Waters Treaty in 1960 to manage river resources. With China, the Panchsheel Agreement of 1954 sought peaceful coexistence, though border disputes remain. Nepal and India share a special bond through the 1950 Treaty of Peace and Friendship. Bhutan’s 1949 treaty, revised in 2007, ensures close cooperation in defense and foreign policy. Bangladesh and India signed the 1972 Treaty of Friendship after its independence. Sri Lanka’s Sirima-Shastri Pact of 1964 addressed the rights of Indian-origin Tamils. Maritime MoUs with Sri Lanka and Maldives highlight India’s oceanic role. Together, geography and treaties define India’s regional diplomacy and security framework."""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## Apply stopwords and filter and then apply stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the list of into sentences
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer=SnowballStemmer('english')
```

```python
## Apply stopwords and filter and then apply Snowball stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the list of into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
## Apply stopwords and filter and then apply Snowball stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the list of into sentences
```

```python
## Apply stopwords and filter and then apply Snowball stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the list of into sentences
```

```python
sentences=nltk.sent_tokenize(paragraph)
```

```python
## Apply stopwords and filter and then apply Snowball stemming
for i in range(len(sentences)):
    # sentences[i]=sentences[i].tolower()
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words) # converting all the list of into sentences
```

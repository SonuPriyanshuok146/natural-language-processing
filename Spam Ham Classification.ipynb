Spam Ham Classification Project Using NLTK

## Using Bag of Words (BOW)

```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.csv', sep=',',names=["label","message"])
```

```python
messages
```

```python
## Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## output features
y=pd.get_dummies(messages['label'])
y=y.iloc[:,0].values
```

```python
## Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
## Create a bag of words models with N Gram
from sklearn.feature_extraction.text import CountVectorizer
## for binary bag of words enable binary = True
cv=CountVectorizer(max_features=200,binary=True,ngram_range=(1,2))
```

```python
len(x_train)
```

```python
## independent features
x_train= cv.fit_transform(x_train).toarray()
x_test=cv.transform(x_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, formatter=dict(float=lambda x: "%.3g" % x))
```

```python
x_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

### TF-IDF

```python
## Train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
## Create the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
x_train=tv.fit_transform(x_train).toarray()
x_test=tv.transform(x_test).toarray()
```

```python
x_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(x_train,y_train)
```

```python
## prediction
y_pred=spam_tfidf_model.predict(x_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

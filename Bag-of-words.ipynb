```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.csv',sep=',',names=["label","message"])
```

```python
messages
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## Create a bag of words models


```python
## Create a bag of words models
from sklearn.feature_extraction.text import CountVectorizer
## for binary bag of words enable binary = True
cv=CountVectorizer(max_features=100,binary=True)
```

```python
x= cv.fit_transform(corpus).toarray()
```

```python
## Create a bag of words models with N Gram
from sklearn.feature_extraction.text import CountVectorizer
## for binary bag of words enable binary = True
cv=CountVectorizer(max_features=200,binary=True,ngram_range=(2,3))
```

```python
x= cv.fit_transform(corpus).toarray()

cv.vocabulary_
```
